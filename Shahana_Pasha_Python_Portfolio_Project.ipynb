```python
import seaborn as sns
mpgdataset = sns.load_dataset("mpg")
mpgdataset.columns = mpgdataset.columns.str.capitalize()
mpgdataset['Origin'] = mpgdataset['Origin'].str.upper()
mpgdataset.rename(columns={'Mpg': 'MPG','Model_year':'Model_Year'},inplace=True)
```

```python
mpgdataset.head(10)
```

```python
# Question 1: Create a visual showing the relationship between horsepower and fuel economy.#
sns.relplot(
    data=mpgdataset,
    x="Horsepower", y="MPG",kind="line")
# As the horsepower increases, the fuel economy(mpg) decreases. #
```

```python
# Question 2: Create a table summarizing fuel economy by country of origin. #
summarytable = mpgdataset.groupby("Origin").MPG.mean().sort_values(ascending=False)
summarytable.to_frame().reset_index()
# Cars made in Japan have the highest fuel economy, followed by Europe and then USA. #
```

```python
# Question 3: Create a visual showing how fuel economy has changed over time. #
sns.relplot(
    data=mpgdataset,
    x="Model_Year", y="MPG",kind="line")
# Fuel economy has increased over the years, peaking in the Model Year 1980. #
```

```python
mpgbycompany = mpgdataset.groupby("Company").MPG.mean().sort_values(ascending=False)
mpgbycompanydata = mpgbycompany.to_frame().reset_index()
mpgbycompanydata
```

```python
sns.catplot(
    data=mpgbycompanydata,
    x="MPG", y="Company",kind="bar",hue="Company",errorbar=None)
# Question 4A : . Create a chart showing fuel economy by Car manufacturers and which company has the highest fuel economy. #
# The chart below shows fuel economy by car manufacturers.#
# Honda has the highest fuel economy, not including a few companies grouped in Other category #
# due to small sample size of data for those particular companies. #
```
